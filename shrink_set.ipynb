```python
import pandas as pd
import numpy as np

giant_file_path = r"C:\Users\Adrian\Downloads\Transportation_Network_Providers_-_Trips_-_2019.csv"
test_file_path = r"C:\Users\Adrian\Downloads\Transportation_Network_Providers_-_Vehicles.csv"
```

```python
def reduce_csv(chunk, columns, reduction_param):
    selected_df = chunk[columns]
    filters = np.random.random((len(selected_df.index), 1))
    return selected_df[(filters < reduction_param )]
```

```python
filter_columns_cars = ["MONTH_REPORTED", "STATE", "MAKE", "MODEL"]
filter_columns_trips = ["Trip ID", "Trip Start Timestamp", "Trip End Timestamp", "Trip Seconds", "Trip Miles", "Trips Pooled", "Pickup Centroid Latitude", "Pickup Centroid Longitude", "Dropoff Centroid Latitude", "Dropoff Centroid Longitude"]


created_df = pd.DataFrame(columns = filter_columns_trips)

chunksize = 10 ** 6
for chunk in pd.read_csv(giant_file_path, chunksize=chunksize):
    temp_df = reduce_csv(chunk, filter_columns_trips, 0.1)
    created_df = pd.concat([created_df, temp_df])

created_df
```
